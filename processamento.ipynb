```python
from nilearn import datasets
```

```python
abide = datasets.fetch_abide_pcp( #baixa a base ejá traz dados pré-processados e guarda os dados na variável abide
    pipeline='cpac',
    band_pass_filtering=True, #Aplica filtro de frequência no sinal cerebral mantém sinais uteis e remove ruidos, imporrtante para conectividade funcional
    global_signal_regression=False, #decide se remove o sinal global do cérebro, nesse casp mantém escolha meotodológica
    derivatives=['func_preproc'], #diz o tipo de dados que eu quero
    n_subjects=100 #limita o número de pessoas
)
```

```python
print(abide) #printa as informações extraidas
```

```python
print(len(abide.func_preproc)) # len(...) → conta quantos elementos existem | abide.func_preproc → lista de exames
```

```python
import pandas as pd # importa a biblioteca a biblioteca pandas e coloca o apelido pd

df = pd.DataFrame(abide.phenotypic) #converte os dados em formato de tabela

print(df.head()) #mostra as primeiras linhas
```

```python
df['label'] = df['DX_GROUP'].apply(lambda x: 1 if x == 1 else 0)
#df['DX_GROUP'] coluna original valores 1 = autismo 2 = controle
#apply aplica a função em cada linha
#lambda x: função pequena (anônima)
#if | else se for autismo =1 senão = 0
```

```python
from nilearn import datasets

atlas = datasets.fetch_atlas_schaefer_2018(n_rois=100)

#um mapa do cerebbro dividido em regiões
#transforma o cérebro em "partes" e cada parte vira uma variável
#Optou-se pelo atlas Schaefer 2018 por sua maior robustez, melhor resolução funcional e melhor compatibilidade com estudos recentes de conectividade cerebral.”
```

```python
from nilearn.input_data import NiftiLabelsMasker

masker = NiftiLabelsMasker(labels_img=atlas.maps, standardize=True)
#NiftiLabelsMasker ferramenta que pega o exame fMRI | extrai sinais de cada região do cérebro
# labels_img=atlas.maps usa o atlas como referência
# standardize=True normaliza os sinais
```

```python
time_series = []

for func_file in abide.func_preproc:
    ts = masker.fit_transform(func_file)
    time_series.append(ts)
# for func_file in ... percorre cada paciente
# fit_transform(...) pega o exame e transforma em matriz: linhas → tempo | colunas → regiões do cérebro
# time_series.append(ts) guarda os dados de cada pessoa
```

```python
from nilearn.connectome import ConnectivityMeasure

connectivity = ConnectivityMeasure(kind='correlation')

conn_matrices = connectivity.fit_transform(time_series)

#calcula correlação entre regiões do cérebro
#gera matriz
#Isso é a conectividade funcional
```

```python
import numpy as np

def flatten_matrix(mat):
    return mat[np.triu_indices_from(mat, k=1)]

X = np.array([flatten_matrix(m) for m in conn_matrices])
y = df['label'].values

#Matriz → vetor Modelos de ML não aceitam matriz 2D
#então transforma em lista de números
# np.triu_indices_from -> pega só metade da matriz (evita duplicação)
# X dados (features)
#y rótulos (ASD ou controle)
```

```python
#VALIDAÇÃO CRUZADA
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

svm = SVC(kernel='linear')

scores = cross_val_score(svm, X, y, cv=5)

print("Acurácias por fold:", scores)
print("Acurácia média:", scores.mean())

#Divide os dados em 5 partes
#Treina e testa várias vezes
#Dá uma média mais confiável
```

```python
#Dividir dados (treino e teste)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2, #20% dos dados vão para teste
    random_state=42, #Garante que o resultado seja sempre igual (reprodutibilidade)
    stratify=y
    )#mantém a proporção de autismo e controle
```

```python
#NORMALIZAÇÃO
#Por que normalizar?
#As features têm escalas diferentes
#Modelos como SVM precisam disso
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

#fit_transform Aprende a escala + aplica (no treino)
#transform Só aplica (no teste)
#Nunca usar fit no teste!
```

```python
#OTIMIZAÇÃO (GridSearch)
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf']
}

grid = GridSearchCV(SVC(), param_grid, cv=5)

grid.fit(X_train, y_train)

print("Melhores parâmetros:", grid.best_params_)
```

```python
#USA O MELHOR MODELO
best_model = grid.best_estimator_
```

```python
#PREVISÃO

y_pred = best_model.predict(X_test)
```

```python
#Avaliação (igual antes):
from sklearn.metrics import accuracy_score, classification_report

print("Acurácia:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#Métrica mais forte (ROC-AUC)
from sklearn.metrics import roc_auc_score

y_prob = best_model.decision_function(X_test)

print("AUC:", roc_auc_score(y_test, y_prob))
```

```python
#Random Forest (comparação)
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)

print("Acurácia RF:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```
